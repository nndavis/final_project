```python
def get_coordinates(city):
    api_link = f"{maps_api}{city}&appid={key}"
    response = requests.get(api_link)
    full_json = response.json()
    city = full_json[0]['name']
    longitude = full_json[0]['lon']
    latitude = full_json[0]['lat']
    coords = (latitude, longitude)
    return coords
```
