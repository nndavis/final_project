```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
reviews_df = pd.read_csv("Resources/glassdoor_reviews.csv")
reviews_df.head()
```

```python
# Create companies dataframe from companies_sorted.csv
companies_df = pd.read_csv("Resources/companies_sorted.csv")
companies_df.head()
```

```python
# Convert firm column to lowercase
reviews_df["firm"].str.lower()
```

```python
# update column name from "firm" to "name"
reviews_df["name"] = reviews_df["firm"].str.lower()
```

```python
df_merge = pd.merge(reviews_df, companies_df, on="name", how="inner")
```

```python
df_merge.head()
```

```python
# create a new dataframe with target columns
df_merge_clean = df_merge[["name" , "job_title" , "industry" , "location" , "overall_rating" , "work_life_balance" , "culture_values" , "diversity_inclusion" , 
"career_opp" , "comp_benefits" , "total employee estimate"]]
```

```python
# drop na values from our dataframe
df_merge_clean = df_merge_clean.dropna()
df_merge_clean.count()
```
